```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("earthquakes_database.csv")
```

```python
df.drop(columns=["Location Source"], inplace=True)
```

```python
df["time"] = pd.to_datetime(df["Time"])
```

```python
df.fillna(0, inplace=True)
```

```python
df["magnitude_squared"] = df["Magnitude"] ** 2  
```

```python
df["magnitude_category"] = pd.cut(df["Magnitude"], bins=[0, 3, 6, 10], labels=["low", "medium", "high"])
```

```python
df["magnitude_category"].value_counts().plot(kind="bar")
plt.title("Frequency of Earthquakes by Magnitude Category")
plt.xlabel("Magnitude Category")
plt.ylabel("Frequency")
plt.show()
```

```python
df.plot(kind="scatter", x="Magnitude", y="Depth", alpha=0.5)
plt.title("Magnitude vs Depth of Earthquakes")
plt.xlabel("Magnitude")
plt.ylabel("Depth")
plt.show()
```

```python
df["magnitude_category"].value_counts().plot(kind="pie", autopct='%1.1f%%')
plt.title("Distribution of Earthquake Magnitudes")
plt.ylabel("")
plt.show()
```

```python
df["Magnitude"].plot(kind="hist", bins=20)
plt.title("Frequency of Earthquakes by Magnitude")
plt.xlabel("Magnitude")
plt.ylabel("Frequency")
plt.show()
```

```python
df.to_csv("cleaned_earthquakes_data.csv", index=False)
```
